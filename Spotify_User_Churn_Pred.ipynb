```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv("spotify_churn_dataset.csv")
data1.head()
```

```python
data1=data1.drop('user_id',axis='columns')
```

```python
data1.head()
```

```python
target=data1['is_churned']
target.head()
```

```python
data1=data1.drop('is_churned',axis='columns')
```

```python
data1.head()
```

// Change to catgorical variables
1.   gender
1.  country
2.   subscription_type
2.  device_type



```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_gender=LabelEncoder()
le_country=LabelEncoder()
le_subscription_type=LabelEncoder()
le_device_type=LabelEncoder()
```

```python
data1['gender_n']=le_gender.fit_transform(data1['gender'])
data1['country_n']=le_country.fit_transform(data1['country'])
data1['subscription_type_n']=le_subscription_type.fit_transform(data1['subscription_type'])
data1['device_type_n']=le_device_type.fit_transform(data1['device_type'])
```

```python
data1.head()
```

```python
data1=data1.drop(['gender','country','subscription_type','device_type'],axis='columns')
```

```python
data1.head()
```

// ALGORITHM
1. LOGISTIC REGRESSION
2. DECISION TREE
3. RANDOM FOREST
4. SVM

```python
x=data1
x.head()
```

```python
y=target
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import linear_model
lg=linear_model.LogisticRegression(solver='liblinear')
lg.fit(x_train,y_train)
lg.score(x_test,y_test)
```

```python
from sklearn import tree
dt=tree.DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
svm=SVC()
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=50)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```
